```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Weekdays(Independent Variable)
x= np.array([1,2,4,6,7]).reshape(-1,1)
# Average Temperature(Dependent Variable)
y=np.array([24,23,21,25,27]).reshape(-1,1)
```

```python
degrees = [1,2,3,4,5]
plt.figure(figsize=(10,6))
```

```python
for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    x_poly = poly_features.fit_transform(x)
    model= LinearRegression()
    model.fit(x_poly, y)
    x_test = np.linspace(1,7,7).reshape(-1,1)
    x_test_poly = poly_features.fit_transform(x_test)
    y_pred = model.predict(x_test_poly)
    plt.plot(x_test, y_pred, label="Degree" + str(degree))
plt.scatter(x,y, label="Temperature Data")
plt.xlabel("WeekDays")
plt.ylabel("Temperature Celsius")
plt.title("Polynomial Regression -Temperature")
plt.legend()
plt.show()
```

```python
x_new = np.array([3,5]).reshape(-1,1)
x_new_poly = poly_features.transform(x_new)
y_new_predict = model.predict(x_new_poly)
print("Predicted temperatures for New WeekDays: ")
for i in range(len(x_new)):
    print("Weekdays", x_new[i], ": Predicted Temperature", y_new_predict[i])
```
